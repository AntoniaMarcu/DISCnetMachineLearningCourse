Generate a bivariate distribution -- zero mean, unit variance of each feature.

```python
N1 = 50000
x = np.random.randn(N1)
y = np.random.randn(N1)
plt.plot(x, y, '.', Color='magenta')
plt.axis('equal')
plt.grid(True)
plt.show()

X = np.matrix([x, y]).T
X.size
X.shape
```

```python
C = np.matrix([[2, 1], [1, 2]])
A = np.linalg.cholesky(C)
print(A)
print(A*A.T)
```

```python
Y = X*A.T
Y.shape
#plt.plot(Y[:,0], Y[:,1], '.', color='blue')
fig, ax =  plt.subplots()
ax.plot(Y[:,0], Y[:,1], '.', color='blue')
ax.set_xlabel('Variable 1')
ax.set_ylabel('Variable 2')
ax.set_title('Scatter of Correlated Variables')
ax.grid('True')
```

```python
th = 0.25;
u = np.matrix([np.cos(th), np.sin(th)]).T
print(u)

su = u.T*C*u
print("Projected Variance:", su)
```

```python
print(np.pi)
thRange = np.linspace(0, 2*np.pi, 50)
N = len(thRange)
print(N)
P1 = np.random.rand(N,1)
P2 = np.random.rand(N,1)
print(P1.shape)

for i in range(0,N):
    th = thRange[i]
    u = np.matrix([np.cos(th), np.sin(th)]).T
    P1[i] = u.T*C*u
    
    C1 = np.cov(Y.T)
    P2[i] = u.T*C1*u
    
    
plt.plot(P1)
plt.plot(P2)

print(C)
print(C1)
```
